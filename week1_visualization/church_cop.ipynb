```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'church_cop.csv'  # Update with your file path if needed
data = pd.read_csv(file_path)

# print(data.head())
```

```python
total_contributions = data['Amount_Contributed'].sum()
members_contributions = data[data['Contributor_Type'] == 'Member']['Amount_Contributed'].sum()
outsiders_contributions = data[data['Contributor_Type'] == 'Outsider']['Amount_Contributed'].sum()

members_percentage = (members_contributions / total_contributions) * 100
outsiders_percentage = (outsiders_contributions / total_contributions) * 100

department_contributions = data[data['Contributor_Type'] == 'Member'].groupby('Department')['Amount_Contributed'].sum()

print(f"Total Contributions: {total_contributions}")
print(f"Members Contributions: {members_contributions} ({members_percentage:.2f}%)")
print(f"Outsiders Contributions: {outsiders_contributions} ({outsiders_percentage:.2f}%)")
print("")
print(department_contributions)
```

```python


department_contributions.plot(kind='bar', color='skyblue', figsize=(8, 5))
plt.title('Contributions by Department')
plt.xlabel('Department')
plt.ylabel('Total Contributions')
plt.xticks(rotation=45)
plt.tight_layout()
# plt.show()
```
